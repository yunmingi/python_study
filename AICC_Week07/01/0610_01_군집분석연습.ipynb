```python
#데이터 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
fruits = np.load('fruits_300.npy')
iris = datasets.load_iris()
```

```python
#100번과 200번의 과일 이미지를 출력 
fig, axs = plt.subplots(1,2)
axs[0].imshow(fruits[100], cmap='gray_r')
axs[1].imshow(fruits[200], cmap='gray_r')
plt.show()
```

```python
#픽셀값 분석 
#사용하기 쉽도록 각 과일 데이터를 나눈다 
apple = fruits[0:100].reshape(-1,100*100)
pineapple = fruits[100:200].reshape(-1,100*100)
banana = fruits[200:300].reshape(-1,100*100)
```

```python
#나누어진 데이터 차원 확인 
apple.shape,pineapple.shape,banana.shape
```

```python
#히스토그램 시각화 
import matplotlib.pyplot as plt

plt.hist(np.mean(apple,axis=1),alpha=0.8)
plt.hist(np.mean(banana, axis=1), alpha=0.8)
plt.hist(np.mean(banana,axis=1),alpha=0.8)
plt.legend(['apple','pineapple','banana'])
plt.show()
```

```python
#서브플롯을 이용해 세 가지 과일에 대한 히스토그램 그리기 
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1,3,figsize=(20,5))
axs[0].set_title('apple')
axs[0].bar(range(10000), np.mean(apple, axis=0))
axs[1].set_title('pineapple')
axs[1].bar(range(10000), np.mean(pineapple, axis=0))
axs[2].set_title('banana')
axs[2].bar(range(10000), np.mean(banana, axis=0))
plt.show()
```

```python
#픽셀 평균값을 100x100크기로 변경하여 이미지처럼 출력하여 그래프와 비교 
#각 과이르이 모든 이미지를 합쳐놓은 개념(대표 이미지)
apple_mean = np.mean(apple, axis=0).reshape(100,100)
pineapple_mean = np.mean(pineapple,axis=0).reshape(100,100)
banana_mean = np.mean(banana, axis=0).reshape(100,100)

fig,axs = plt.subplots(1,3,figsize=(20,5))
axs[0].imshow(apple_mean, cmap='gray_r')
axs[1].imshow(pineapple_mean,cmap='gray_r')
axs[2].imshow(banana_mean, cmap='gray_r')
plt.show()
```

```python
# fruits배열에 있는 샘플들의 절대값의 평균 구하기
# 샘플들의 절대값 구하기
abs_diff = np.abs(fruits - apple_mean)
abs_diff[:3]
```

```python
# 절대값의 평균 구하기
abs_mean = np.mean(abs_diff, axis=(1,2)) # axis축. 면(0), 행(1), 열(2)
abs_mean[:3]
```

```python
# 결과 배열의 크기 확인
abs_mean.shape
```

```python
for i in range(10):
    for j in range(10):
        print(i, j, i*10 + j)
```

```python
# 절대값 평균을 가장 작은 순서대로 100개를 가져옴(apple_mean과 오차가 가장 작은 샘플 100개를 고르는 것)

# abs_mean에서 오름차순으로 나열한 abs_mean인덱스를 반환
apple_index = np.argsort(abs_mean)[:100]
apple_index
```

```python
apple_index.shape
```

```python
# 선택된 인덱스에 해당하는 샘플을 10 x 10 격자로 그래프 그리기
fig, axs = plt.subplots(10, 10, figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[apple_index[i*10 + j]], cmap='gray_r')
        #axs[i, j].axis('off')
plt.show()
```

```python
#### 바나나 사진의 평균 banana_mean과 비슷한 사진 100장을 찾아 출력하기
abs_diff = np.abs(fruits - banana_mean)
abs_mean = np.mean(abs_diff, axis=(1,2))

banana_index = np.argsort(abs_mean)[:100]
fig, axs = plt.subplots(10, 10, figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[banana_index[i*10 + j]], cmap='gray_r')
        axs[i, j].axis('off')
plt.show()
```

```python
import pandas as pd
from sklearn import datasets

# 붓꽃 데이터셋 로드
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 산점도 행렬
sns.pairplot(df, hue='species', diag_kind='kde')
plt.show()
```

```python
from sklearn.cluster import KMeans

# 데이터셋에서 특성만 선택
X = df.iloc[:, :-1]

# K-평균 군집화
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)
```

```python
# 결과 시각화
sns.pairplot(df, hue='cluster', diag_kind='kde')
plt.show()
```
