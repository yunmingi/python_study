```python
from urllib.parse import urljoin
```

```python
url_base = "https://www.wanted.co.kr/"
url_sub = "wdlist/518?country=kr&job_sort=job.recommend_order&years=-1&selected=1025&selected=1024&selected=655&selected=1634&selected=669&locations=all"
url = url_base + url_sub
```

```python
soup_list = soup.find_all('div', 'JobCard_JobCard__oZL4d')
soup_list
```

```python
# for문으로  url 데이터 리스트로 추출
url_add = []
for item in soup_list:
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

```python
# 포문으로 돌면서 소재지 및 경력 사항 추가
location = []
career = []

driver = webdriver.Chrome()

for idx, row in df.iterrows():
    # url 컬럼의 0번부터 끝까지 순서대로 접속
    url = row['URL']
    driver.get(url)
    soup_tmp = BeautifulSoup(driver.page_source, 'html.parser')

    # 접속한 url에서 필요한 정보 찾기
    element = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[1]/div/section/header/div/div[1]').text.split('∙')
    # 해당 정보 빈 리스트에 저장
    location.append(element[1])
    career.append(element[2])
    # 진행도 확인하기 위해 idx 출력
    print(idx)
```
