```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import pandas as pd

def get_data_from_current_page(driver):
    elements = driver.find_elements(By.CLASS_NAME, "list_body")
    url_list = []
    for element in elements:
        strong_tags = element.find_elements(By.CLASS_NAME, 'str_tit')
        for strong_tag in strong_tags:
            url_list.append(strong_tag.get_attribute("href"))
    return url_list

# 해당 페이지로 이동하는 함수
def click_page_by_number(driver, page_number):
    try:
        xpath = f'//*[@id="default_list_wrap"]/div/button[{page_number}]'
        page_button = driver.find_element(By.XPATH, xpath)
        page_button.click()
        print(f"{page_number} 페이지 클릭")
        return True
    except NoSuchElementException:
        print(f"{page_number} 페이지를 찾을 수 없습니다.")
        return False

# 첫 번째 페이지부터 시작
url_list = []

# 현재 페이지에서 데이터 수집
current_data = get_data_from_current_page(driver)
url_list.extend(current_data)
# 수집한 데이터 출력
print(f"현재 페이지 데이터 수집 완료. 수집된 데이터 수: {len(url_list)}")

# 1부터 3까지의 페이지 클릭하면서 데이터 수집
for page_number in range(1, 9):
    click_page_by_number(driver, page_number)
    # 현재 페이지에서 데이터 수집
    current_data = get_data_from_current_page(driver)
    url_list.extend(current_data)
    # 수집한 데이터 출력
    print(f"{page_number} 페이지 데이터 수집 완료. 수집된 데이터 수: {len(url_list)}")

# 수집한 URL 정보 출력
df = pd.DataFrame({'URL': url_list})
print(df)
```

```python
# Selenium 웹 드라이버 시작
driver = webdriver.Chrome()  # 여기에서 Chrome() 객체를 생성합니다.

company = []
title = []
career = []
education = []
employee_form = []
payment = []
keyword_tag = []
location = []

url = df['URL']
# URL을 하나씩 순회하며 처리합니다.
for i in url:
    try:
        # 각 URL로 이동
        driver.get(i)
        
        # 페이지 로드 대기
        time.sleep(2)
        
        # 페이지 소스를 BeautifulSoup로 파싱
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # 필요한 요소를 찾아 텍스트를 가져옵니다.
        try:
            title_select = driver.find_element(By.CLASS_NAME, "tit_job").text
        except:
            title_select = None
        
        try:
            company_list = driver.find_element(By.XPATH, '//*[@id="content"]/div[3]/section[1]/div[1]/div[1]/div/div[1]/a[1]').text
        except:
            company_list = None
            
        try:
            payment_list = driver.find_element(By.XPATH, '//*[@id="content"]/div[3]/section[1]/div[1]/div[2]/div/div[2]/dl[1]/dd').text
        except:
            payment_list = None
        
        try:
            location_raw = driver.find_element(By.XPATH, '//*[@id="content"]/div[3]/section[1]/div[1]/div[2]/div/div[2]/dl[2]/dd').text
        except:
            location_raw = None
        
        try:
            career_select = driver.find_element(By.XPATH, '//*[@id="content"]/div[3]/section[1]/div[1]/div[2]/div/div[1]/dl[1]/dd/strong').text
        except:
            career_select = None
        
        try:
            education_list = driver.find_element(By.XPATH, '//*[@id="content"]/div[3]/section[1]/div[1]/div[2]/div/div[1]/dl[2]/dd/strong').text
        except:
            education_list = None
        
        # 데이터를 리스트에 추가합니다.
        
        title.append(title_select)
        company.append(company_list)
        career.append(career_select)
        payment.append(payment_list)
        location.append(location_raw)
        education.append(education_list)
    
    except Exception as e:
        print(f"Error processing URL {url}: {e}")

# 결과를 데이터프레임으로 변환합니다.
result_df = pd.DataFrame({
    'Title': title,
    'Company': company,
    'Career': career,
    'Payment': payment,
    'Location': location,
    'Education': education
    #'Urls': urls
})

# 결과 출력
print(result_df)
```
